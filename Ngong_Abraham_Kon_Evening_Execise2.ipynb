1.	Write a Pandas program to compare the elements of two Pandas series. 
    Sample Series [4,65,436,3,9], [7,0,3,897,9]

```python
SER_1 = pd.Series([4, 65, 436, 3, 9])
SER_2 = pd.Series([7, 0, 3, 897, 9])

result = SER_1 == SER_2
print(result)
```

2.	Write a Pandas program to add, subtract, multiply and divide two Pandas series. 
Sample Series: [2,4,6,8,14], [1,3,5,7,9]


```python
SER1 = pd.Series([2, 4, 6, 8, 14])
SER2 = pd.Series([1, 3, 5, 7, 9])

addition = SER1 + SER2
subtraction = SER1 - SER2
multiplication = SER1 * SER2
division = SER1 / SER2

print("Addition:")
print(addition)
print("\nSubtraction:")
print(subtraction)
print("\nMultiplication:")
print(multiplication)
print("\nDivision:")
print(division)
```

4.	Write a Pandas program to convert a given series to an array.

```python
series_1 = pd.Series(['Love', 800, 'Joy', 789.9, 'Peace', True])

array_1 = np.array(series_1)
print(array_1)
```

5.	Write a Pandas program to display the most frequent value in the given series and replace everything else as ‘Other’ in the series. (Use the ‘HomeTeamGoals’ column in the dataset provided)

```python
# loading dataset here to check for HomeTeamGoals column
DF = pd.read_csv('AfricaCupofNationsMatches.csv')

series = pd.Series(DF.HomeTeamGoals)

most_frequent_value = series.mode()[0]
series = series.map(lambda x: most_frequent_value if x == most_frequent_value else 'Other')

print(series)
```
