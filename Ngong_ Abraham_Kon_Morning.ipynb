# BELOW USING "UceResultsBySchool2011-2016.csv"

```python
# importing required libraries

import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
import os # this is operating system to interact with jupyter notebook
pd.options.display.max_columns=50
```

```python
# loading csv files into dataframe
df = pd.read_csv('UceResultsBySchool2011-2016.csv')
df
```

```python
df.head(10)
```

```python
df.tail(10)
```

```python
# checking for the shape of the dataset
df.shape
```

```python
# checking for the columns of the dataset
df.columns
```

```python
# getting some basic information about dataset
df.info()
```

```python
# describing the dataset using transpose method where T is transpose
df.describe().T
```

```python
# checking for null values
null = df.isnull().sum().sort_values(ascending = True)
null
```

```python
# checking null values in figure
import missingno as msno
msno.bar(df,color="yellow",fontsize=14)
plt.show()
```

```python
# removing rows with null values
dropna = df.dropna()
dropna
```

```python
# checking for duplicate values in the dataset
duplicated = df.duplicated().sum()
duplicated
```

```python
# filling rows with null values with empty space
df.fillna(' ',inplace = True)
df
```

```python
# plotting Histogram for years in the dataset
df['YEAR'].plot(kind='hist',title ='GRAPH SHOWING YEARS FROM 2011 to 2016')
plt.xlabel('Years')
plt.show()
```

```python
# plotting Histogram for female candidates in the dataset
df['FEMALE CANDIDATES'].plot(kind='hist',title = 'GRAPH OF FEMALE CANDIDATE')
plt.xlabel('Females')
plt.show()
```

```python
# plotting Histogram for male candidates in the dataset
df['MALE CANDIDATES'].plot(kind='hist',title = 'GRAPH OF MALE CANDIDATES')
plt.xlabel('Males')
plt.show()
```

```python
corr = df.corr()
```

```python
ax = sns.heatmap(corr,cbar =1 ,linewidths =2,vmax =1, vmin =0.9,square=True,cmap ='Blues')
plt.title('Heatmap UCE Results From 2011 -2016')
plt.show()
```
