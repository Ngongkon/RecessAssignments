# Analyzing Data / Cleaning Data Exercise

```python
#loading mine.csv file
df1=pd.read_csv('mine.csv')
df1
```

# Using head()method to get first 6 rows

```python
df1.head(6)
```

# Using tail()method to get last 6 rows

```python
df1.tail(6)
```

# info() method show the summary of the dataset

```python
df1.info()
```

# isnull()is used to check for null values

```python
df1.isnull().sum().sort_values(ascending=False)
```

# dropna() is used to remove rows or columns with missing values or NaN

```python
df1.dropna()
```

# fillna() Fills all missing values with a specific value

```python
df1=df1.fillna(0)
df1
```

# Remove rows with Duration, Pulse, Maxpulse & Calories that are not an int. Reduce rows from 169 to 164

```python
df1 = df1.loc[pd.to_numeric(df1['Duration'], errors='coerce').notnull()] 
df1 = df1.loc[pd.to_numeric(df1['Pulse'], errors='coerce').notnull()] 
df1 = df1.loc[pd.to_numeric(df1['Maxpulse'], errors='coerce').notnull()] 
df1 = df1.loc[pd.to_numeric(df1['Calories'], errors='coerce').notnull()]
df1
```

# duplicated() Returns a boolean Series indicating duplicate rows

```python

df1.duplicated()
```

# drop_duplicates() Removes duplicate rows

```python

df1=df1.drop_duplicates()
df1
```

# corr()  Calculates the correlation between columns using different such as kendall etc

```python

df1.corr(method='kendall')
```
